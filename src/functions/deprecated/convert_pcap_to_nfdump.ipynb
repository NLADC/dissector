```python
import os
import subprocess
from multiprocessing import Pool

p = Pool(8)
to_parse = []
directory = "/mnt/d"

def parse_file(filename):
    filename_full = os.path.join(directory, filename)
    print filename + " started" 
    outfolder = directory + "/nfdump/" + filename + "/"
    os.mkdir(outfolder)
    p = subprocess.Popen(["functions/nfdump/bin/nfpcapd -r " + filename_full + " -l " + outfolder], shell=True, stdout=subprocess.PIPE)
    file_info, error = p.communicate()
    p.wait()
    # print "functions/nfdump/bin/nfpcapd -r " + filename + " -l " + directory + "/temp"
    print filename + " finished"

for filename in os.listdir(directory):
    if filename.endswith(".pcap"): 
        to_parse.append(filename)
        
print to_parse
print p.map(parse_file, to_parse)
```
